```python
import numpy as np 
import pandas as pd 
from sklearn.impute import SimpleImputer

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Criar um dataframe de exemplo com valores ausentes
np.random.seed(42) # 42 é a taxa de repetição dos numeros

#Gaussiana: Parâmetro 0 é a média da distribuição normal. 
# # O parâmetro 1 é desvio padrão na distribuição normal.
# O 100 é o número de valores a serem criados. Serão gerados 100 valores aleatórios para cada coluna a partir de uma distribuição normal. 
data = {'A' : np.random.normal(0,1,100),
        'B' : np.random.normal(0,1,100)} 

df = pd.DataFrame(data)

df.iloc[10:20,0] = np.nan # definir valores ausentes na coluna A nas linhas de 10 a 19. 
df.iloc[30:40,1] = np.nan # este terceiro parâmetro é o número da coluna. No caso, como inicia-se de 0, esta seria a coluna B;

# Visualizar os padrões ausentes

sns.heatmap(df.isnull(), cbar=False, cmap='crest') #cmap - cor
plt.show()
```
